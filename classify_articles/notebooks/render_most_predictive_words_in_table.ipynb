```python
import pandas as pd
```

```python
# A small bit of code to render the most predictive words and accompanying probabilites for each desk to use in blog post
arrays = [['Arts', 'Arts', 'Business', 'Business', 'Obituaries', 'Obituaries', 'Sports', 'Sports', 'World', 'World'],
         ['Word', 'Predictive Power', 'Word', 'Predictive Power','Word', 'Predictive Power','Word', 'Predictive Power','Word', 'Predictive Power',]]
```

```python
# read the most predictive words json file output from NYT classifier code
df = pd.read_json("/Users/cameroncairns/code/NYT_article_classification/classify_articles/results/most_discriminating_words.json")
```

```python
tuples = list(zip(*arrays))
index = pd.MultiIndex.from_tuples(tuples, names=['Desk', ''])
df_results = pd.DataFrame([], columns=index)
```

```python
for idx, row in df.iterrows():
    for desk in ['Arts', 'Business', 'Obituaries', 'Sports', 'World']:
        df_results.loc[idx, (desk, "Word")] = row.loc[desk][0]
        df_results.loc[idx, (desk, "Predictive Power")] = row.loc[desk][1]
```
